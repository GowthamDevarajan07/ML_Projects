```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('movies.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
selected_features = ['genres','keywords','vote_average','cast', 'director']
print(selected_features)
```

```python
for feature in selected_features:
  df[feature] = df[feature].fillna(' ')
```

```python
df[['genres','keywords','vote_average','cast', 'director']].isnull().sum()
```

```python
# now we ll combine all the selectected in features in one column
combined_features = df['genres']+' '+df['keywords']+' '+df['vote_average'].astype('str')+' '+df['cast']+' '+df['director']
```

```python
combined_features.shape
```

```python
print(combined_features)
```

```python
#converting the text data to feature vectors

vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

Cosine Similarity

```python
#getting the similarity score using cosine similarity

similarity_score = cosine_similarity(feature_vectors)
similarity_score
```

```python
similarity_score.shape
```

Here each movie is compared with all the movies and similariy score is given , tellling how much a movie is similar to the movie

```python
movie_name = input('Enter a movie name ')
```

```python
#creating a list with all the movie names
all_movies = list(df['title'])
all_movies
```

```python
close_match = difflib.get_close_matches(movie_name,all_movies)
print(close_match)
```

```python
# taking the most closest match
closest_match = close_match[0]
closest_match
```

```python
df.columns
```

```python
index_of_movie = df[df.title==closest_match]['index'].values[0] # if we give just values it gives an array
index_of_movie
```

```python
# getting a similar movie

similarity_score_with_ironman = list(enumerate(similarity_score[index_of_movie]))

#enumerate each value with index in the format ('index','item')

similarity_score_with_ironman
```

```python
len(similarity_score_with_ironman)
```

```python
# sorting the movies based on similarity score

sorted_similar_movies = sorted(similarity_score_with_ironman, key=lambda x:x[1], reverse = True)
sorted_similar_movies
```

```python
# Print name of the similar movies based on the index

print('Movies suggested for the user \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = df[df['index']==index]['title'].values[0]
  if(i<31):
    print(i,'.',title_from_index)
    i+=1
```

```python
#Movie Recommendation System

movie_name = input('Enter a movie name ')
all_movies = list(df['title'])
close_match = difflib.get_close_matches(movie_name,all_movies)
closest_match = close_match[0]
index_of_movie = df[df.title==closest_match]['index'].values[0]
similarity_score_with_userinput_movie = list(enumerate(similarity_score[index_of_movie]))
sorted_similar_movies = sorted(similarity_score_with_userinput_movie, key=lambda x:x[1], reverse = True)

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = df[df['index']==index]['title'].values[0]
  if(i<31):
    print(i,'.',title_from_index)
    i+=1
```
